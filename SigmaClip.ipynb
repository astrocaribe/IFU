Sigma Clipping Test for ifu_3d_collapse() Function
==================================================

This notebook is a test of the sigma clipping function of the ifu_3d_collapse() function.  
The issue turned out to be that the numpy.median() function was not calculating the median  
correctly, and ultimately found one that did work, numpy.ma.extras.median(). I might have  
to investigate this issue and submit a pull request eventually.

Proof of Concept
================

Here, we experiment with the different pieces to fully test the sigma 
clipping functionality.

```python
# Setup inline plotting
%matplotlib inline
```

Create a datacube with a random sampling given mu and sigma, and display a collapsed median across the entire z-axis...

```python
# Import required modules
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from __future__ import print_function
import numpy as np

import sys
sys.path.append('./scripts/')

# Import the IFU scripts
from ifu_math import *
from ifu_3d_collapse import *

# Generate a cube of random numbers
mu, sigma = 0, 15
cube = np.random.normal(loc=mu, scale=sigma, size=(5, 5, 5))

# Collapse the sample with a median
collapsed = ifu_3d_collapse(cube, method='median')

# Display center frame
fig, ax = plt.subplots(figsize=(6, 6))
cax = ax.imshow(collapsed, cmap=cm.hot, interpolation='nearest')
fig.colorbar(cax, shrink=.8)

# Store the color limits for later use...
cxLims = cax.get_clim()

# Diagnostic messages...
print()
print('Datacube min({}), max({}), median({})'
      .format(cube.min(), cube.max(), np.median(cube)))
print('Collapsed min({}), max({}), median({})'
      .format(collapsed.min(), collapsed.max(), np.median(collapsed)))
```

Now that we have a random datacube with a "first estimate" collapse. Now let's 
look at the sigma clipping portion ...

**Note**: The `sigma_clip()` function returns a numpy.ma array! This just means that 
the output has to be checked carefully, especially when used in additional 
calculations as required for these tools in development. The dashed `--` entries 
below correspond to the clipped values.

```python
from astropy.stats.funcs import sigma_clip

# Sigma clip the same sample for comparison
sigma = 1.3
clipped = sigma_clip(cube, sigma, iters=None, axis=0, copy=True)

# Print the clipped array (numpy.ma masked array)
print('Clipped values (type numpy.ma):')
print(clipped)
print()

# Retrieve/print the good values
print('Good data points (z, y, x):')
good = clipped.nonzero()
good
```

```python
# Comparing two ways to retrieve the good datapoints for verification...
print(clipped.data[~clipped.mask])
print()
print(cube[good])
```

```python
# Sigma clip the same sample for comparison
clipped_3d = ifu_3d_collapse(cube, method='median', sigma=1.5)
print(clipped_3d)
print()
print(collapsed)
```
