Sigma Clipping Test for ifu_3d_collapse() Function
==================================================

This notebook is a test of the sigma clipping function of the ifu_3d_collapse() function.  
The issue turned out to be that the numpy.median() function was not calculating the median  
correctly, and ultimately found one that did work, numpy.ma.extras.median(). I might have  
to investigate this issue and submit a pull request eventually.

1. Proof of Concept (Image)
===========================

Here, we experiment with the different pieces to fully test the sigma 
clipping functionality. We'll start by first calculating sigma clipped  
samples of an image to verify the output of `astropy.stats.funcs.sigma_clip()`. 
The final goal is to compare the result of collapsing a datacube using 
astropy vs. using `ifu_3d_collapse()`.

```python
# Setup inline plotting environment
%matplotlib inline
```

```python
# Import global modules
import numpy as np
from astropy.stats.funcs import sigma_clip
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from __future__ import print_function
```

### 1.1 Generate a Random Sample
Generate a random Gaussian distribution given $\mu$ and $\sigma$. Revise the min/max values in the output below to get a feel for the range generated.

```python
# Generate an image (2D) of random numbers given mu, sigma
mu, sigma = 0, 15
image = np.random.normal(loc=mu, scale=sigma, size=(20, 20))

# Display the image w/ a colorbar
sample = image
fig, (ax, a_hist) = plt.subplots(1, 2, figsize=(18, 6))
cax = ax.imshow(sample, cmap=cm.hot, interpolation='nearest')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$\mu$={}, $\sigma$={}, sample={}'.format(mu, sigma, image.size))
fig.colorbar(cax, ax=ax)

# Retrieve image plotting limits for later use
im_min, im_max = cax.get_clim()


# Set the number of bins based on binsize
sample_min, sample_max = np.min(sample), np.max(sample)
binsize = 1
num_bins = np.int(np.floor((sample_max - sample_min) / binsize))

print('{} bins needed based on input binsize of {}.'.format(num_bins, binsize))
print('Sample min/max: [{} / {}]'.format(sample_min, sample_max))

# Plot a histogram of the sample
n, bins, patches = a_hist.hist(sample.flatten(), num_bins, lw=2., color='green')
a_hist.set_xlabel('Sample')
a_hist.set_ylabel('Frequency')
a_hist.set_title('Full Sample')

# Get histogram limits to include below
hist_lim = a_hist.get_xlim()
```

### 1.2 Set Sigma Clipping to 3$\sigma$ or Above
What happens if we clip above 3$\sigma$? 3$\sigma$ includes ~ 99% in a Gaussian distribution, so going > 3$\sigma$ should clip **No** values from the original sample...

```python
# ****************************** Clip Sample (3σ+)******************************
# Perform a sigma clip, given a sigma limit. > 3 should return the same array
clipped_sample = sigma_clip(sample, sig=3.5, iters=None, copy=True)

# Display clipped sample image
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 9))
cax1 = ax1.imshow(sample, cmap=cm.hot, interpolation='nearest')
ax1.set_title('Original Sample')
fig.colorbar(cax1, ax=ax1, shrink=0.73)

cax2 = ax2.imshow(clipped_sample, cmap=cm.hot, interpolation='nearest', 
                  vmin=im_min, vmax=im_max)
ax2.set_title('Sigma Clipped Sample')
fig.colorbar(cax2, ax=ax2, shrink=0.73)

# Generate a new array without the clipped values...
clipped_array = clipped_sample.compressed()
        
# Check for sample/clipped sample parity...
if np.array_equal(sample.flatten(), clipped_array): 
    print('Arrays are exactly the same, no clipping occured!')
else:
    print('Clipping occured.')
# ****************************** Clip Sample (3σ+)******************************
```

Good!

### 1.3 Clip a Substantial Amount From the Sample, ~ 1$\sigma$
Let's perform the analysis then with a hugely clipped sample...

```python
# ****************************** Clip Sample (1σ)******************************
# Perform a sigma clip, given a sigma limit. > 3 should return the same array
clipped_sample = sigma_clip(sample, sig=1, iters=None, copy=True)

# Display clipped sample image
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 9))
cax1 = ax1.imshow(sample, cmap=cm.hot, interpolation='nearest')
ax1.set_title('Original Sample')
fig.colorbar(cax1, ax=ax1, shrink=0.73)

cax2 = ax2.imshow(clipped_sample, cmap=cm.hot, interpolation='nearest', 
                  vmin=im_min, vmax=im_max)
ax2.set_title('Sigma Clipped Sample')
fig.colorbar(cax2, ax=ax2, shrink=0.73)

# Generate a new array without the clipped values...
clipped_array = clipped_sample.compressed()
        
# Check for sample/clipped sample parity...
if np.array_equal(sample.flatten(), clipped_array): 
    print('Arrays are exactly the same, no clipping occured!')
else:
    print('Clipping occured.')
# ****************************** Clip Sample (1σ)******************************
    

# ****************************** Clipped Histogram (1σ)******************************
    # Display the clipped image w/ a colorbar
fig, (ax, a_hist) = plt.subplots(1, 2, figsize=(18, 6))
cax = ax.imshow(clipped_sample.filled(fill_value=np.nan), cmap=cm.hot, 
                interpolation='nearest', vmin=im_min, vmax=im_max)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('$\mu$={}, $\sigma$={}, good data={}'.format(mu, sigma, clipped_array.size))
fig.colorbar(cax, ax=ax)

# Set the number of bins based on binsize
sample_min, sample_max = np.min(clipped_array), np.max(clipped_array)
binsize = binsize
num_bins = np.int(np.floor((sample_max - sample_min) / binsize))

print('{} bins needed based on input binsize of {}.'.format(num_bins, binsize))
print('Sample min/max: [{} / {}]'.format(sample_min, sample_max))

# Plot a histogram of the sample
n, bins, patches = a_hist.hist(clipped_array, num_bins, lw=2., color='cyan')
a_hist.set_xlim(hist_lim)
a_hist.set_xlabel('Sample')
a_hist.set_ylabel('Frequency')
a_hist.set_title('Clipped Sample')
# ****************************** Clipped Histogram (1σ)******************************
```

### 1.4 Selecting/displaying good/bad data...

```python
good = clipped_sample.data[~clipped_sample.mask]
bad = clipped_sample.data[clipped_sample.mask]

print('Retained Data, {} items:'.format(len(good)))
print(good)
print()
print('Rejected Data, {} items:'.format(len(bad)))
print(bad)
```

... and let's display the first line in the original, clipped, and boosted samples ...

```python
test = clipped_sample.nonzero()
boosted = np.copy(sample)
boosted[test] = boosted[test]*100

print('Original Data:')
print(clipped_sample[0, :].data)
print()
print('Clipped Data:')
print(clipped_sample[0, :])
print()
print('Boosted Data (only to non-clipped):')
print(boosted[0, :])
print()
```

So in essence, the original data has been boosted by 100., but only for the 'good' data, or the unclipped data in this example.

```python
# Import required modules
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from __future__ import print_function
import numpy as np

import sys
sys.path.append('./scripts/')

# Import the IFU scripts
from ifu_math import *
from ifu_3d_collapse import *

# Generate a cube of random numbers
mu, sigma = 0, 15
cube = np.random.normal(loc=mu, scale=sigma, size=(5, 5, 5))

# Collapse the sample with a median
collapsed = ifu_3d_collapse(cube, method='median')

# Display center frame
fig, ax = plt.subplots(figsize=(6, 6))
cax = ax.imshow(collapsed, cmap=cm.hot, interpolation='nearest')
fig.colorbar(cax, shrink=.8)

# Store the color limits for later use...
cxLims = cax.get_clim()

# Diagnostic messages...
print()
print('Datacube min({}), max({}), median({})'
      .format(cube.min(), cube.max(), np.median(cube)))
print('Collapsed min({}), max({}), median({})'
      .format(collapsed.min(), collapsed.max(), np.median(collapsed)))
```
